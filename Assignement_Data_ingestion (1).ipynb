```python
def fetch_fhir_data(resource, count=50, max_pages=3):
    import requests
    from datetime import datetime, timedelta
    #url = f"https://hapi.fhir.org/baseR4/{resource}?_count={count}"
    #all_data = []
    #page = 0
    

    last_run = (datetime.now() - timedelta(days=2)).strftime("%Y-%m-%d")

    url = f"https://hapi.fhir.org/baseR4/{resource}?_count={count}&_lastUpdated=gt{last_run}"
    all_data = []
    page = 0
    while url and page < max_pages:
        res = requests.get(url).json()

        if "entry" in res:
            for item in res["entry"]:
                if "resource" in item:
                    all_data.append(item)

        # next page
        next_url = None
        for link in res.get("link", []):
            if link["relation"] == "next":
                next_url = link["url"]

        url = next_url
        page += 1

    return all_data
```

```python
from delta.tables import DeltaTable

gold = DeltaTable.forName(spark, "gold_patient")

gold.alias("target").merge(
    df_silver.alias("source"),
    "target.patient_id = source.patient_id AND target.is_current = true"
).whenMatchedUpdate(
    condition="""
        target.gender <> source.gender OR
        target.birthDate <> source.birthDate OR
        target.family_name <> source.family_name OR
        target.given_name <> source.given_name
    """,
    set={
        "is_current": "false",
        "end_date": "current_timestamp()"
    }
).whenNotMatchedInsert(
    values={
        "patient_id": "source.patient_id",
        "gender": "source.gender",
        "birthDate": "source.birthDate",
        "family_name": "source.family_name",
        "given_name": "source.given_name",
        "is_current": "true",
        "start_date": "current_timestamp()",
        "end_date": "NULL"
    }
).execute()
```
